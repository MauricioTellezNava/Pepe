# Análisis preliminar del shooting list de Pepe la papa

```python
import pandas as pd
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/drive/MyDrive/Pepe la papa/Shooting list.xlsx")
df.head()
```

```python
df.columns
```

## Capitulos

```python
df["capitulo"].value_counts()
```

```python
conteo = df["capitulo"].value_counts().sort_values()
etiquetas = df["capitulo"].unique()
colores = ["#EE6055", "#60D394", "#AAF683", "#FFD97D", "#FF9B85"]
extraccion = 0,0,0.2

fig, ax = plt.subplots()
plt.pie(conteo, 
        labels= etiquetas,
        autopct = ("%0.2f %%"),
        colors = colores,
        explode = extraccion, 
        shadow=True)

ax.set_title("Porcentaje de tomas por capítulo \n")
ax.legend()
plt.show()
```
